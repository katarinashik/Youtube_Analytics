```python
import pandas as pd
import numpy as np
import re

def clean_date(date_str):
    # Check if the input is a string
    if isinstance(date_str, str):
        # Remove ordinal suffixes (st, nd, rd, th) using regex
        date_str = re.sub(r'(\d+)(st|nd|rd|th)', r'\1', date_str)
    return date_str

df = pd.read_csv('/content/top_20_enriched (1).csv')
df['Created Date'] = df['Created Date'].apply(clean_date)
df['Created Date'] = pd.to_datetime(df['Created Date'], format='%b %d, %Y', errors='coerce')
# Added errors='coerce' to handle invalid parsing gracefully
```
