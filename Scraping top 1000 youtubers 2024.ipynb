# Scrapping a list of top Youtubers in 2024 from  https://us.youtubers.me/global/all/top-1000-most-subscribed-youtube-channels

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL for the webpage i need to scrap
url = "https://us.youtubers.me/global/all/top-1000-most-subscribed-youtube-channels"

# Send a GET request to the URL to fetch the webpage content
response = requests.get(url)

# there are a lot of special characters in the names of channels, so i use utf-8 encoding to see them properly
response.encoding = 'utf-8'
```

```python
# parse the HTML content of the webpage using BeautifulSoup with the HTML parser
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())
```

```python
# find the table with the top ytbers and extract all table rows ("tr") from the table
lines = soup.find("table", class_ = "top-charts").find_all("tr")
lines
```

```python
# empty list to store the names of YouTube channels
names = []

# Iterate through all rows except the first one (header row) to extract channel names
for line in lines[1:]:

    #placeholder of a channel name
    name = line.find("img")['alt']
    #adding names to my list
    names.append(name)

names
```

```python
#convert to dataframe
df = pd.DataFrame(names, columns=["Channel Name"])
df
```
