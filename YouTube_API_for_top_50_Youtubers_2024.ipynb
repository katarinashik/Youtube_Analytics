# Searching info on the top youtubers from the list with YT API

```python
# YouTube API endpoints and API key

SEARCH_URL = "https://www.googleapis.com/youtube/v3/search"
CHANNELS_URL = "https://www.googleapis.com/youtube/v3/channels"
API_KEY = ""

# List of top channels, the code will be done multiple times with different lists because of the youtube API quota here is the first 10
liste = [
  'MrBeast',
  'T-Series',
  'Cocomelon - Nursery Rhymes',
  'YouTube Movies',
  'SET India',
  'Vlad and Niki',
  '✿ Kids Diana Show',
  'Like Nastya',
  'Music',
  'Zee Music Company',
]
# Placeholder to store raw API responses to be sure i'm not out of my daily quota
responses=[]

# empty list to store results for processed data
results = []

# loop through each channel in the list
for channel in liste:
    search_params = {
        'part': 'snippet',
        'q': f'{channel}',
        'type': 'channel',
        'maxResults': 1,
        'key': API_KEY
    }

    # Make a request to the search endpoint
    response = requests.get(SEARCH_URL, params=search_params)
    # Parse the response as JSON
    search_data = response.json()

    # extract channel id with search to pass it to channel endpoint
    if "items" in search_data and len(search_data["items"]) > 0:
        item = search_data["items"][0]
        channel_id = item["id"]["channelId"]

      #search detailed channel information via channel ID
        channel_params = {
            'part': 'snippet,statistics',
            'id': channel_id,
            'key': API_KEY
        }
        # Make a request to the channels endpoint
        channel_response = requests.get(CHANNELS_URL, params=channel_params)
        # Parse the response as JSON
        channel_data = channel_response.json()

        #acces to the data

    if "items" in channel_data and len(channel_data["items"]) > 0:
                item = channel_data["items"][0] #  Take the first item
                snippet = item.get("snippet", {}) # Extract the snippet data
                statistics = item.get("statistics", {}) # Extract the statistics data

                # Collect channel details into a dictionary
                channel_details = {
                    "Title": snippet.get("title", "N/A"),
                    "Country": snippet.get("country", "N/A"),
                    "Subscribers": statistics.get("subscriberCount", "N/A"),
                    "Video Views": statistics.get("viewCount", "N/A"),
                    "Uploads": statistics.get("videoCount", "N/A"),
                    "Date of Creation": snippet.get("publishedAt", "N/A")
                }
                results.append(channel_details)  # add the details to the results list


results10 = results
results10
```
